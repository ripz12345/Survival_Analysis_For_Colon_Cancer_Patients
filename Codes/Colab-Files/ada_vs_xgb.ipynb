```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
```

```python
from google.colab import drive
drive.mount('/content/drive')
# importing our CSV file into a pandas dataframe
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dataset/colon.csv")
```

```python
mean_value_nodes=df['nodes'].mean()
df['nodes'].fillna(value=mean_value_nodes, inplace=True)

mean_value_differ=df['differ'].mean()
df['differ'].fillna(value=mean_value_differ, inplace=True)
```

```python
df['nodes'] = df['nodes'].astype(int)
df['differ'] = df['differ'].astype(int)
```

```python
# getting independent variables
X=df[['rx','sex','age','obstruct','perfor','adhere','nodes','differ','extent','surg','time']]
```

```python
y=df[['status']]
```

```python
def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]==y_hat[i]==0:
           TN += 1
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1

    return(TP, FP, TN, FN)
```

```python
for i in range(100):
  X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size = 0.2)
  rfr = RandomForestRegressor(n_estimators = 100)
  knn = KNeighborsRegressor(n_neighbors=20, metric='euclidean')
  ada = AdaBoostRegressor() #
  ada.fit(X_train1,y_train1)
  xgb = XGBRegressor() #
  xgb.fit(X_train1,y_train1)
  y_pred1 = xgb.predict(X_test1)
  y_test1 = np.asarray(y_test1)
  a=np.where(y_test1<0.5,0,1)
  b=np.where(y_pred1<0.5,0,1)
  tp1,fp1,tn1,fn1=perf_measure(a,b)
  acc1=(tp1+tn1)/(tp1+fp1+tn1+fn1)
  print("Accuracy: :",acc1)
  f = open("/content/drive/MyDrive/Colab Notebooks/accuXGB.txt","a+") #
  acc=str(format(acc1, '.4f'))
  acc=acc+'\n'
  f.writelines(acc)
  f.close()
```
