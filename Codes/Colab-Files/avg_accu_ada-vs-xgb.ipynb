```python
fADA = open("/content/drive/MyDrive/Colab Notebooks/accuADA.txt","r")
fXGB = open("/content/drive/MyDrive/Colab Notebooks/accuXGB.txt","r")
```

```python
Lines = fADA.readlines()
arrADA=[]
count = 0
# Strips the newline character
for line in Lines:
  count += 1
  arrADA.append(float(line.strip()))
print(arrADA)
sumADA = 0
for i in range(len(arrADA)):
  sumADA += arrADA[i]
avgADA = sumADA/len(arrADA)
avgADA = format(avgADA, '.4f')
print("avgADA = ",avgADA)
```

```python
Lines = fXGB.readlines()
arrXGB=[]
count = 0
# Strips the newline character
for line in Lines:
  count += 1
  arrXGB.append(float(line.strip()))
print(arrXGB)
sumXGB = 0
for i in range(len(arrXGB)):
  sumXGB += arrXGB[i]
avgXGB = sumXGB/len(arrXGB)
avgXGB = format(avgXGB, '.4f')
print("avgXGB = ",avgXGB)
```

```python
import matplotlib.pyplot as plt
import math
avg_ADA = 'avgADA ' + avgADA
avg_XGB = 'avgXGB ' + avgXGB
data = {avg_ADA:float(avgADA), avg_XGB:float(avgXGB)}
keys = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (7, 5))
 
# creating the bar plot


plt.xlabel("Base estimators used")
plt.ylabel("Average accuarcy")
plt.title("Performance comparison measurement")
max_ylim = 0.95
min_ylim = 0.90
plt.ylim(min_ylim, max_ylim)
plot = plt.bar(keys, values, color =['maroon','blue','green'], width = 0.3)
plt.show()
```
