```python
fDT = open("/content/drive/MyDrive/Colab Notebooks/accuDT.txt","r")
fRFR = open("/content/drive/MyDrive/Colab Notebooks/accuRFR.txt","r")
fKNN = open("/content/drive/MyDrive/Colab Notebooks/accuKNN.txt","r")
```

```python
Lines = fDT.readlines()
arrDT=[]
count = 0
# Strips the newline character
for line in Lines:
  count += 1
  arrDT.append(float(line.strip()))
print(arrDT)
sumDT = 0
for i in range(len(arrDT)):
  sumDT += arrDT[i]
avgDT = sumDT/len(arrDT)
avgDT = format(avgDT, '.4f')
print("avgDT = ",avgDT)
```

```python
Lines = fKNN.readlines()
arrKNN=[]
count = 0
# Strips the newline character
for line in Lines:
  count += 1
  arrKNN.append(float(line.strip()))
print(arrKNN)
sumKNN = 0
for i in range(len(arrKNN)):
  sumKNN += arrKNN[i]
avgKNN = sumKNN/len(arrKNN)
avgKNN = format(avgKNN, '.4f')
print("avgKNN = ",avgKNN)
```

```python
Lines = fRFR.readlines()
arrRFR=[]
count = 0
# Strips the newline character
for line in Lines:
  count += 1
  arrRFR.append(float(line.strip()))
print(arrRFR)
sumRFR = 0
for i in range(len(arrRFR)):
  sumRFR += arrRFR[i]
avgRFR = sumRFR/len(arrRFR)
avgRFR = format(avgRFR, '.4f')
print("avgRFR = ",avgRFR)
```

```python
import matplotlib.pyplot as plt
import math
avg_DT = 'ADA DT ' + str(avgDT)
avg_RFR = 'ADA RFR ' + str(avgRFR)
avg_KNN = 'ADA KNN ' + str(avgKNN)
data = {avg_DT:float(avgDT), avg_RFR:float(avgRFR), avg_KNN:float(avgKNN)}
keys = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (7, 5))
 
# creating the bar plot


plt.xlabel("Base Estimators used for AdaBoost regression--->")
plt.ylabel("Average accuarcy --->")
plt.title("Performance comparison between AdaBoost regression with different type of Base Estimators")
max_ylim = 0.95
min_ylim = 0.90
plt.ylim(min_ylim, max_ylim)
plot = plt.bar(keys, values, color =['maroon','blue','green'], width = 0.3)
plt.show()
```
