```python
fXGB = open("/content/drive/MyDrive/Colab Notebooks/accuXGB_PM.txt","r")
```

```python
Lines = fXGB.readlines()
arrXGB=[]
count = 0
# Strips the newline character
for line in Lines:
  count += 1
  arrXGB.append(float(line.strip()))
print(arrXGB)
sumXGB = 0
for i in range(len(arrXGB)):
  sumXGB += arrXGB[i]
avgXGB = sumXGB/len(arrXGB)
avgXGB = format(avgXGB, '.4f')
print("avgXGB = ",avgXGB)
```
